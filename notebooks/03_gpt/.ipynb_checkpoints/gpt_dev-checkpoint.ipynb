```python
# Initial setup
torch.manual_seed(2134789)

project_root = Path.cwd().parents[1]
print(project_root)

device = 'cuda' if torch.cuda.is_available() else 'cpu'
print(device)
```

```python
with open(project_root/'data/tiny_shakespeare.txt', 'r', encoding='utf-8') as f:
    text = f.read()
```

```python
chars = sorted(list(set(text)))
vocab_size = len(chars)

print("Vocab size:", vocab_size)
print(''.join(chars))
```

```python
stoi = {s: i for i, s in enumerate(chars)}
itos = {i: s for s, i in stoi.items()}

print(stoi)
print(itos)
```

```python
encode = lambda s: [stoi[c] for c in s]
decode = lambda l: ''.join([itos[i] for i in l])


print(encode("Hi, Friendo"))
print(decode([20, 47, 6, 1, 18, 56, 47, 43, 52, 42, 53]))
```

```python
# Encode the entire text dataset
data = torch.tensor(encode(text), dtype=torch.long)
print(data.shape, data.dtype)
print(data[:100])
```

```python
n = int(0.9 * len(data))
train_data = data[:n]
val_data = data[n:]

print(len(train_data))
print(len(val_data))
```

```python
block_size = 8
x = train_data[:block_size]
y = train_data[1: block_size+1]
for t in range(block_size):
    context = x[:t+1]
    target = x[t]
    print(f"{context} --> {target}") # For a block size of block_size we get block_size training examples 
```

```python
batch_size = 4
block_size = 8

def get_batch(split, batch_size, block_size):
    data = train_data if split == "train" else val_data
    ix = torch.randint(0, len(data) - block_size, (batch_size, ))
    xb = torch.stack([data[i: i+block_size] for i in ix])
    yb = torch.stack([data[i+1:i+block_size+1] for i in ix])
    return xb, yb


xb, yb = get_batch("train", batch_size, block_size) # Our input to the transformer

xb, yb # 32 Indenpendent examples
```

## Bigram -- (Baseline)

```python
class BigramLanguageModel(nn.Module):
    def __init__(self, vocab_size):
        super().__init__()
        self.lookup_table = nn.Embedding(vocab_size, vocab_size)

    def forward(self, idx, targets=None):
         logits = self.lookup_table(idx) # (B, T, C)

         if targets is None:
            loss = None
         else:   
            B, T, C = logits.shape
            loss = F.cross_entropy(
                logits.view(B*T, C),
                targets.view(B*T)
            )
            
         return logits, loss

    def generate(self, idx, max_new_tokens):
        for _ in range(max_new_tokens):
            logits, _ = self(idx)  # (B, T, C)
            logits = logits[:, -1, :] # (B, C)
            probs = F.softmax(logits, dim=-1) # (B, C)
            idx_next = torch.multinomial(probs, num_samples=1) # (B, 1)
            idx = torch.cat((idx, idx_next), dim=1) # (B, T+1)
        return idx

```

```python
bigram_model = BigramLanguageModel(vocab_size).to(device)
optimizer = torch.optim.AdamW(bigram_model.parameters(), lr=1e-3)
```

```python
# Dummy Forward Pass
logits, loss = bigram_model(xb.to(device))
logits.shape, loss
```

```python
a = torch.tensor(5, device=device)
type(a.detach())
```

```python
lossi = []
max_steps = 20000
batch_size = 32
block_size = 8
for i in range(max_steps):
    xb, yb = get_batch("train", batch_size, block_size)
    logits, loss = bigram_model(xb.to(device), yb.to(device))
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    lossi.append(loss.detach())
    if i % 1000 == 0 or i == max_steps-1:
        avg_loss = torch.stack(lossi[-1000:]).mean().item()
        print(f"Step: {i:5d} : Loss: {avg_loss: .4f}")
```

```python
plt.plot(torch.stack(lossi).view(-1, 100).mean(1).cpu())
plt.title("Train Loss Curve (Smoothed)")
plt.xlabel("Steps")
plt.ylabel("Loss")
```

```python
W = bigram_model.lookup_table.weight.detach().cpu().numpy()
plt.figure(figsize=(8, 8))
plt.imshow(W, cmap='viridis')
plt.colorbar()
plt.xlabel('output / next-token logits')
plt.ylabel('input token (row index)')
plt.title('Bigram embedding (logit) table')
plt.show()
```

## Transformer

```python
# Mathematical  trick in self-attention
import torch
B, T, C = 4, 8, 2
x = torch.randn(B, T, C)
x.shape # Currently the 8 tokens are not talking to each other
# The easiest way to just average over the previous tokens (Do not include futute tokens)
```

```python
xbow = torch.zeros((B, T, C))
for b in range(B):
    for t in range(T):
        xprev = x[b,:t+1] # (t, C)
        xbow[b, t] = torch.mean(xprev, 0)
```
